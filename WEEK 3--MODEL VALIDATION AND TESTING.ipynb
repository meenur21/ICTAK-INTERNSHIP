# **IDENTIDYING OUTLIERS**

```python
from scipy import stats

#function to detect outliers using Z-score
def detect_outliers_zscore(column):
    z_scores = stats.zscore(column)
    return (z_scores > 3) | (z_scores < -3)


for column in credit.select_dtypes(include=['float64']).columns:
    outliers = detect_outliers_zscore(credit[column])
    print(f"Outliers in column '{column}': {credit[outliers][column]}")

```

## Visualising the outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

def show_outliers(data, col, edge):
    # Create a figure with two subplots side by side
    fig, axs = plt.subplots(1, 2, figsize=(6, 6))

    # Plot data points where column values are less than the threshold
    sns.stripplot(y=col, data=data[data[col] < edge], ax=axs[0])
    axs[0].set_title(f"{col} < {edge}")

    # Plot data points where column values are greater than the threshold
    sns.stripplot(y=col, data=data[data[col] > edge], ax=axs[1])
    axs[1].set_title(f"{col} > {edge}")

    # Adjust layout for better spacing
    plt.tight_layout()

    # Show the plots
    plt.show()
```

## **Scaling Numerical Columns and encoding categories**

```python
X_train.columns
```

```python
X_train.dtypes
```

***One hot encoder and robust scaler are used for encoding and scaling respectively***

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define the numerical columns (excluding one-hot encoded columns)
numerical_cols = ['Annual_Income', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Monthly_Inhand_Salary', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Num_Credit_Inquiries', 'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month', 'Amount_invested_monthly', 'Monthly_Balance']

# Define the categorical columns for one-hot encoding
categorical_cols = ['Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour', 'Occupation']

# Define ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numerical_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ],
    remainder='passthrough'  # keep remaining columns as they are
)

# Fit and transform the training data
X_train_processed = preprocessor.fit_transform(X_train)

# Transform the validation and testing data
X_val_processed = preprocessor.transform(X_val)
X_test_processed = preprocessor.transform(X_test)

# Get the feature names after one-hot encoding
feature_names = numerical_cols + list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols))

# Convert the transformed data into DataFrame with column names
X_train_processed_df = pd.DataFrame(X_train_processed, columns=feature_names)
X_val_processed_df = pd.DataFrame(X_val_processed, columns=feature_names)
X_test_processed_df = pd.DataFrame(X_test_processed, columns=feature_names)
```

# **Developing Classification Models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

# Create the classifiers
logistic_regression = LogisticRegression()
decision_tree = DecisionTreeClassifier()
random_forest = RandomForestClassifier()
gradient_boosting = GradientBoostingClassifier()
svm = SVC()

# Train the classifiers
logistic_regression.fit(X_train_processed_df, y_train)
decision_tree.fit(X_train_processed_df, y_train)
random_forest.fit(X_train_processed_df, y_train)
gradient_boosting.fit(X_train_processed_df, y_train)
svm.fit(X_train_processed_df, y_train)
```

# **Evaluation of Models On Validation Set**

```python
from sklearn.metrics import accuracy_score

# Predictions on the validation set
y_pred_logistic_regression = logistic_regression.predict(X_val_processed_df)
y_pred_decision_tree = decision_tree.predict(X_val_processed_df)
y_pred_random_forest = random_forest.predict(X_val_processed_df)
y_pred_gradient_boosting = gradient_boosting.predict(X_val_processed_df)
y_pred_svm = svm.predict(X_val_processed_df)

# Accuracy scores on the validation set
accuracy_logistic_regression = accuracy_score(y_val, y_pred_logistic_regression)
accuracy_decision_tree = accuracy_score(y_val, y_pred_decision_tree)
accuracy_random_forest = accuracy_score(y_val, y_pred_random_forest)
accuracy_gradient_boosting = accuracy_score(y_val, y_pred_gradient_boosting)
accuracy_svm = accuracy_score(y_val, y_pred_svm)

# Print accuracy scores
print("Validation Accuracy (Logistic Regression):", accuracy_logistic_regression)
print("Validation Accuracy (Decision Tree):", accuracy_decision_tree)
print("Validation Accuracy (Random Forest):", accuracy_random_forest)
print("Validation Accuracy (Gradient Boosting):", accuracy_gradient_boosting)
print("Validation Accuracy (SVM):", accuracy_svm)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Precision scores
precision_logistic_regression = precision_score(y_val, y_pred_logistic_regression, average='macro')
precision_decision_tree = precision_score(y_val, y_pred_decision_tree, average='macro')
precision_random_forest = precision_score(y_val, y_pred_random_forest, average='macro')
precision_gradient_boosting = precision_score(y_val, y_pred_gradient_boosting, average='macro')
precision_svm = precision_score(y_val, y_pred_svm, average='macro')

# Recall scores
recall_logistic_regression = recall_score(y_val, y_pred_logistic_regression, average='macro')
recall_decision_tree = recall_score(y_val, y_pred_decision_tree, average='macro')
recall_random_forest = recall_score(y_val, y_pred_random_forest, average='macro')
recall_gradient_boosting = recall_score(y_val, y_pred_gradient_boosting, average='macro')
recall_svm = recall_score(y_val, y_pred_svm, average='macro')

# F1-scores
f1_logistic_regression = f1_score(y_val, y_pred_logistic_regression, average='macro')
f1_decision_tree = f1_score(y_val, y_pred_decision_tree, average='macro')
f1_random_forest = f1_score(y_val, y_pred_random_forest, average='macro')
f1_gradient_boosting = f1_score(y_val, y_pred_gradient_boosting, average='macro')
f1_svm = f1_score(y_val, y_pred_svm, average='macro')

# Print the scores
print("Precision (Logistic Regression):", precision_logistic_regression)
print("Precision (Decision Tree):", precision_decision_tree)
print("Precision (Random Forest):", precision_random_forest)
print("Precision (Gradient Boosting):", precision_gradient_boosting)
print("Precision (SVM):", precision_svm)

print("\nRecall (Logistic Regression):", recall_logistic_regression)
print("Recall (Decision Tree):", recall_decision_tree)
print("Recall (Random Forest):", recall_random_forest)
print("Recall (Gradient Boosting):", recall_gradient_boosting)
print("Recall (SVM):", recall_svm)

print("\nF1-score (Logistic Regression):", f1_logistic_regression)
print("F1-score (Decision Tree):", f1_decision_tree)
print("F1-score (Random Forest):", f1_random_forest)
print("F1-score (Gradient Boosting):", f1_gradient_boosting)
print("F1-score (SVM):", f1_svm)
```

***After evaluating various metrics, the Random Forest Classifier has been identified as the best model for the dataset. To further assess its performance and optimize its parameters, the next steps involve conducting evaluations and tuning the model.***

```python
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train_processed_df, y_train)

feature_importances = rf_classifier.feature_importances_
feature_names = X_train_processed_df.columns

# Sorting feature importances in descending order
sorted_indices = feature_importances.argsort()[::-1]


plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_importances)), feature_importances[sorted_indices], align='center')
plt.xticks(range(len(feature_importances)), [feature_names[i] for i in sorted_indices], rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance Score')
plt.title('Feature Importance Plot (Before Hyperparameter Tuning)')
plt.show()
```

```python
import numpy as np

# Calculate cumulative importance
cumulative_importance = np.cumsum(feature_importances[sorted_indices])

# Plot feature importances
plt.figure(figsize=(20, 6))

# Plot individual feature importance
plt.subplot(1, 2, 1)
plt.bar(range(len(feature_importances)), feature_importances[sorted_indices], align='center')
plt.xticks(range(len(feature_importances)), [feature_names[i] for i in sorted_indices], rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance Score')
plt.title('Feature Importance Plot (Before Hyperparameter Tuning)')

# Plot cumulative importance
plt.subplot(1, 2, 2)
plt.plot(range(len(cumulative_importance)), cumulative_importance, marker='o', linestyle='-')
plt.xlabel('Number of Features')
plt.ylabel('Cumulative Importance')
plt.title('Cumulative Feature Importance Plot (Before Hyperparameter Tuning)')

plt.tight_layout()
plt.show()
```

# **Hyperparameter Tuning**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [None, 10, 20],  # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4]  # Minimum number of samples required at each leaf node
}

# Initialize RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=3, scoring='accuracy')

# Perform grid search
grid_search.fit(X_train_processed_df, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Get the best estimator
best_estimator = grid_search.best_estimator_

# Evaluate the best model on validation set
val_accuracy = best_estimator.score(X_val_processed_df, y_val)
print("Validation Accuracy:", val_accuracy)
```

# ***Cross Validation***

---



## K-FOLD Validation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rf_model = RandomForestClassifier(max_depth=None, min_samples_leaf=1, min_samples_split=2, n_estimators=300, random_state=42)

# Perform 10-fold cross-validation
cv_scores = cross_val_score(rf_model, X_train_processed_df, y_train, cv=10)

# Print the results
print("Accuracy scores for each fold:", cv_scores)
print("Average cross-validation accuracy: {:.4f}%".format(cv_scores.mean() * 100))
print("Standard deviation of CV scores: {:.4f}".format(cv_scores.std()))
```

## STRATIFIED VALIDATION

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Setup StratifiedKFold
skf = StratifiedKFold(n_splits=5)

# Perform Stratified Cross-Validation
cv_scores = cross_val_score(rf_model, X_train_processed_df, y_train, cv=skf)

# Print the detailed results
print("Detailed accuracy scores for each fold:", cv_scores)
print("Average cross-validation accuracy: {:.4f}%".format(cv_scores.mean() * 100))
print("Standard deviation of CV scores: {:.4f}".format(cv_scores.std()))
```

```python
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_validate

scoring = {'accuracy': make_scorer(accuracy_score),
           'precision': make_scorer(precision_score, average='macro'),
           'recall': make_scorer(recall_score, average='macro'),
           'f1': make_scorer(f1_score, average='macro')}

scores = cross_validate(rf_model, X_train_processed_df, y_train, cv=5, scoring=scoring)

print("Accuracy: ", scores['test_accuracy'].mean())
print("Precision: ", scores['test_precision'].mean())
print("Recall: ", scores['test_recall'].mean())
print("F1 Score: ", scores['test_f1'].mean())
```

# **Threshold Selection**

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.preprocessing import label_binarize

# Binarize the labels for multi-class evaluation
y_test_binarized = label_binarize(y_test, classes=np.unique(y_test))

# Predict probabilities for each class
probabilities = rf_model.predict_proba(X_test_processed_df)

# Calculate average precision and plot PR curves for each class
precision_dict = {}
recall_dict = {}
average_precision_dict = {}
for i, class_label in enumerate(np.unique(y_test)):
    precision_dict[class_label], recall_dict[class_label], _ = precision_recall_curve(y_test_binarized[:, i], probabilities[:, i])
    average_precision_dict[class_label] = average_precision_score(y_test_binarized[:, i], probabilities[:, i])

    plt.plot(recall_dict[class_label], precision_dict[class_label], lw=2, label=f'Class {class_label} (AP = {average_precision_dict[class_label]:0.2f})')

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend(loc="best")
plt.title("Precision-Recall curve per class")
plt.show()
```

```python
# Calculate micro-average Precision-Recall curve and average precision
precision_micro, recall_micro, _ = precision_recall_curve(y_test_binarized.ravel(), probabilities.ravel())
average_precision_micro = average_precision_score(y_test_binarized, probabilities, average="micro")

print(f"Micro-average precision-recall score: {average_precision_micro:.2f}")

# Optional: Plot micro-average Precision-Recall curve
plt.plot(recall_micro, precision_micro, color='gold', lw=2, label=f'Micro-average PR curve (area = {average_precision_micro:0.2f})')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend(loc="best")
plt.title("Micro-Average Precision-Recall")
plt.show()
```
